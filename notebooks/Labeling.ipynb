```python
import pandas as pd
import numpy as np
import logging
from pathlib import Path

df = pd.read_csv('df_stance_training_data.csv')
```

```python
from datasets import Dataset
df['label'] = df['label'].astype(int)
dataset = Dataset.from_pandas(df)
dataset_split = dataset.train_test_split(test_size=0.2, seed=42)
train_ds = dataset_split['train']
test_ds = dataset_split['test']

print(f"\nEntrenamiento: {len(train_ds)} ejemplos")
print(f"Prueba (Validaci√≥n): {len(test_ds)} ejemplos")
```

```python
from setfit import SetFitModel, SetFitTrainer
from sentence_transformers.losses import CosineSimilarityLoss

# 1. Descargar el modelo base (Multiling√ºe y r√°pido)
model_id = "dccuchile/bert-base-spanish-wwm-uncased"
print(f"Descargando modelo base: {model_id}...")

model = SetFitModel.from_pretrained(model_id)

# 2. Configurar el Entrenador
# - num_epochs: 1 suele bastar para SetFit (evita overfitting)
# - num_iterations: 20 genera 20 pares de comparaci√≥n por cada texto (clave para aprender r√°pido)
trainer = SetFitTrainer(
    model=model,
    train_dataset=train_ds,
    eval_dataset=test_ds,
    loss_class=CosineSimilarityLoss,
    metric="accuracy",
    batch_size=4,
    num_iterations=5, 
    num_epochs=1,
    column_mapping={"text": "text", "label": "label"} 
)

# 3. ¬°Entrenar!
# (Esto tomar√° entre 2 a 10 minutos dependiendo de tu PC)
print("üöÄ Iniciando Fine-Tuning... (Ve por un caf√© r√°pido)")
trainer.train()

# 4. Evaluar qu√© tal qued√≥
metrics = trainer.evaluate()
print("\n" + "="*30)
print(f"üéØ ACCURACY FINAL: {metrics['accuracy']:.2%}")
print("="*30)
```

```python
import shutil

# 1. Definir ruta
path_modelo = MODEL_DIR / "mi_modelo_stance_chileno"

# (Opcional) Limpiar la carpeta si ya exist√≠a para evitar errores
try:
    shutil.rmtree(path_modelo)
except FileNotFoundError:
    pass

# 2. Guardar
model.save_pretrained(path_modelo)
print(f"‚úÖ Modelo guardado en: {path_modelo}")

# --- PRUEBA DE FUEGO INMEDIATA ---
# Probemos con frases que NO estaban en tu entrenamiento

frases_prueba = [
    "Es inaceptable que sigamos postergando esta necesidad urgente de las familias.",  # Deber√≠a ser: En Contra (del retraso) o A Favor (del proyecto) dependiendo contexto
    "Valoro la iniciativa, sin embargo, el financiamiento es irresponsable.",         # Deber√≠a ser: En Contra
    "Este proyecto viene a solucionar un vac√≠o legal hist√≥rico.",                     # Deber√≠a ser: A Favor
    "Solicito que se oficie al ministerio para recabar m√°s antecedentes."             # Deber√≠a ser: Neutral
]

print("\n--- Probando el modelo nuevo ---")
preds = model.predict(frases_prueba)

mapa_labels = {0: "En Contra", 1: "Neutral", 2: "A Favor"}

for frase, pred in zip(frases_prueba, preds):
    # Convertir tensor/numpy a int
    pred_int = int(pred)
    print(f"Texto: {frase[:60]}...")
    print(f"Predicci√≥n: {mapa_labels[pred_int]}\n")
```
